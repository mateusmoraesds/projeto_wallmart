```
Lojas_consolidadas = spark.sql("""
                              SELECT
                                'Kaggle_API'              AS ORIGEM_DADO_CAT_LOJA,
                                'CSV'                     AS ORIGEM_DADO_LISTA_LOJA,
                                C.ID_Store                AS ID_STORE,
                                C.Store                   AS STORE,
                                L.Address                 AS ADDRESS,
                                L.City                    AS CITY,
                                L.State                   AS STATE,
                                C.Type                    AS TYPE,
                                C.Size                    AS SIZE,
                                L.ingestion_date          AS ATUALIZADO_EM
                              FROM Categorias AS C
                              LEFT JOIN Lista_Lojas as L
                                ON C.ID_Store = L.ID_Store
                            """)
display(Lojas_consolidadas)
```

```
delta_path = "/mnt/datalake/3_3_2_lojas_consolidado_silver/delta/"
Lojas_consolidadas.write.mode("overwrite").format("parquet").saveAsTable("3_3_2_lojas_consolidado_silver")
Lojas_consolidadas.write.mode("overwrite").format("delta").save(delta_path)

# Opcional: criar uma tabela Delta no catálogo para facilitar consultas
spark.sql(f"CREATE TABLE IF NOT EXISTS 3_3_2_lojas_consolidado_silver USING DELTA LOCATION '{delta_path}'")
```
